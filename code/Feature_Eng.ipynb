# AutoEncoder

```python
import numpy as np
import pandas as pd
import seaborn as sns
from tqdm import tqdm
import matplotlib.pyplot as plt
import copy
import time

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler

from sklearn.preprocessing import MinMaxScaler    
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight
```

```python
!cp /content/drive/MyDrive/MLSD/hw1/data/tfidf_em_data.json /content/
```

```python
df = pd.read_json("/content/tfidf_em_data.json")
df
```

```python
color_df = pd.get_dummies(df['color']).to_numpy()
cpu_df = pd.get_dummies(df['CPU']).to_numpy()

X_df = []
for title, sub_title, desc, price, size, year, room_number, mileage, color, cpu, ram in zip(
    df['title_encoding'], 
    df['sub_title_encoding'], 
    df['description_encoding'], 
    df['price'], 
    df['size'], 
    df['year'], 
    df['room_number'], 
    df['mileage'], 
    color_df, 
    cpu_df, 
    df['RAM']):
  
    tmp = title
    tmp.extend(sub_title)
    tmp.extend(desc)
    tmp.extend([price])
    tmp.extend([size])
    tmp.extend([year])
    tmp.extend([room_number])
    tmp.extend([mileage])
    tmp.extend(color)
    tmp.extend(cpu)
    tmp.extend([ram])
    X_df.append(np.array(tmp).astype('float32'))
X_df = pd.DataFrame(X_df)
X_df
```

```python
X = {'train': None, 'test': None, 'val': None}
X_trainval, X['test'] = train_test_split(X_df, test_size=0.2, shuffle=True, random_state=21)
X['train'], X['val'] = train_test_split(X_trainval, test_size=0.1, shuffle=True, random_state=42)
```

```python
scaler = MinMaxScaler()
X['train'] = scaler.fit_transform(X['train'])
X['val'] = scaler.transform(X['val'])
X['test'] = scaler.transform(X['test'])
X['train'] = np.array(X['train'])
X['val'] = np.array(X['val'])
X['test'] = np.array(X['test'])
```

```python
class Data(Dataset):
    def __init__(self, x_train):
        self.x = x_train
        self.len = self.x.shape[0]

    def __getitem__(self,index):      
        return self.x[index]

    def __len__(self):
        return self.len
```

```python
batch_size = 256
weighted_loss = False
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

dataloaders = {}
dataset_sizes = {}

for phase in ['train', 'val', 'test']:
    data_set = Data(X[phase])
    dataset_sizes[phase] = data_set.len
    dataloaders[phase] = DataLoader(dataset=data_set, batch_size=batch_size)
```

```python
class MyAutoEncoder(nn.Module):
    def __init__(self, in_dim, latent_dim=16):
        super(MyAutoEncoder, self).__init__()

        self.dim = [8192, 4096, 2048, 512, 256, 128]

        self.encoder = nn.Sequential(
            nn.Linear(in_dim, self.dim[0]),
            # nn.BatchNorm1d(self.dim[0]),
            nn.Dropout(0.5),
            nn.ReLU(),
            nn.Linear(self.dim[0], self.dim[1]),
            # nn.BatchNorm1d(self.dim[1]),
            nn.Dropout(0.4),
            nn.ReLU(),
            nn.Linear(self.dim[1], self.dim[2]),
            # nn.BatchNorm1d(self.dim[2]),
            nn.Dropout(0.3),
            nn.ReLU(),
            nn.Linear(self.dim[2], self.dim[3]),
            # nn.BatchNorm1d(self.dim[3]),
            nn.Dropout(0.2),
            nn.ReLU(),
            nn.Linear(self.dim[3], self.dim[4]),
            # nn.BatchNorm1d(self.dim[4]),
            nn.Dropout(0.1),
            nn.ReLU(),
            nn.Linear(self.dim[4], self.dim[5]),
            # nn.BatchNorm1d(self.dim[4]),
            nn.Dropout(0.1),
            nn.ReLU(),
            nn.Linear(self.dim[5], latent_dim)
        )

        self.bn = nn.BatchNorm1d(latent_dim).to(device)
        self.drop = nn.Dropout(0.05).to(device)
        self.relu = nn.ReLU().to(device)

        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, self.dim[5]),
            # nn.BatchNorm1d(self.dim[4]),
            nn.Dropout(0.1),
            nn.ReLU(),
            nn.Linear(self.dim[5], self.dim[4]),
            # nn.BatchNorm1d(self.dim[4]),
            nn.Dropout(0.1),
            nn.ReLU(),
            nn.Linear(self.dim[4], self.dim[3]),
            # nn.BatchNorm1d(self.dim[3]),
            nn.Dropout(0.2),
            nn.ReLU(),
            nn.Linear(self.dim[3], self.dim[2]),
            # nn.BatchNorm1d(self.dim[2]),
            nn.Dropout(0.3),
            nn.ReLU(),
            nn.Linear(self.dim[2], self.dim[1]),
            # nn.BatchNorm1d(self.dim[1]),
            nn.Dropout(0.4),
            nn.ReLU(),
            nn.Linear(self.dim[1], self.dim[0]),
            # nn.BatchNorm1d(self.dim[0]),
            nn.Dropout(0.5),
            nn.ReLU(),
            nn.Linear(self.dim[0], in_dim)
        )
    

    def forward(self, x):
        latent = self.encoder(x)
        # x = self.bn(latent)
        x = self.drop(latent)
        x = self.relu(x)
        out = self.decoder(x)

        return out, latent
```

```python
model = MyAutoEncoder(in_dim=X['train'].shape[1], latent_dim=16)
model = model.to(device)

criterion = nn.CrossEntropyLoss()

learning_rate=3e-5
optimizer=torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
def train_model(model, criterion, optimizer, num_epochs):
    since = time.time()

    best_model_wts = copy.deepcopy(model.state_dict())
    best_loss = 500.0
    loss_history = {'train': [], 'val': []}

    for epoch in range(num_epochs):
        print(f'Epoch {epoch+1}/{num_epochs}')
        print('-' * 10)

        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()
            else:
                model.eval()

            running_loss = 0.0

            for inputs in dataloaders[phase]:
                inputs = inputs.to(device)
                optimizer.zero_grad()
                with torch.set_grad_enabled(phase == 'train'):
                    outputs, _ = model(inputs)
                    loss = criterion(outputs, inputs)

                    running_loss += loss.item()
                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

            running_loss = running_loss/dataset_sizes[phase]
            print(f'{phase} Loss: {running_loss:.6f}')
            
            loss_history[phase].append(running_loss)

            # deep copy the model
            diff = 1e-5
            if phase == 'val' and running_loss+diff < best_loss:
                best_loss = running_loss
                best_model_wts = copy.deepcopy(model.state_dict())
                print('** Model Saved **')

        print()

    time_elapsed = time.time() - since
    print(f'Training complete in {time_elapsed // 60:.0f}m {time_elapsed % 60:.0f}s')

    # load best model weights
    model.load_state_dict(best_model_wts)
    return model, loss_history
```

```python
model, loss_history = train_model(model, criterion, optimizer, num_epochs=300)
```

```python
# plt.figure(figsize=(20,10))
plt.plot(np.log(loss_history['train']), label="train")
plt.plot(np.log(loss_history['val']), label="val")
plt.title('Reconstruction')
plt.legend()
plt.show()
```

```python
model.eval()
```

```python
test_loss = 0.0
for inputs in dataloaders['test']:
    with torch.no_grad():
        inputs = inputs.to(device)
        outputs, _ = model(inputs)
        test_loss += criterion(outputs, inputs).item()
test_loss = test_loss/dataset_sizes['test']
print(f"Test Loss: {test_loss:.6f}")
```
